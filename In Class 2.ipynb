```python
import numpy as np
import matplotlib.pyplot as plt
import math

N = 1000

mean1 = [1, 0]
mean2 = [-1, 0]
mean3 = [0, -10]

U = np.array([[1/math.sqrt(2), -1/math.sqrt(2)],[1/math.sqrt(2), 1/math.sqrt(2)]])

lambda1 = np.array([[4, 0],[0, 1]])
lambda2 = np.array([[1, 0],[0, 5]])

cov1 = np.matmul(np.matmul(U, lambda1), U.T)
cov2 = np.matmul(np.matmul(U, lambda2), U.T)
cov3 = np.eye(2)

data1 = np.random.multivariate_normal(mean1, cov1, N)
data2 = np.random.multivariate_normal(mean2, cov2, N)
data3 = np.random.multivariate_normal(mean3, cov3, N)



plt.plot(data1.T[0], data1.T[1], '.r', data2.T[0], data2.T[1], '.g', data3.T[0], data3.T[1], '.b')
plt.show()
```

```python
class1 = np.array([[],[]])
class2 = np.array([[],[]])
class3 = np.array([[],[]])


for i in data1:
    rand = int(np.random.rand()*3)
    i = np.array([i]).T
    if rand == 0:
        class1 = np.hstack((class1, i))
    if rand == 1:
        class2 = np.hstack((class2, i))
    if rand == 2:
        class3 = np.hstack((class3, i))
        
for i in data2:
    rand = int(np.random.rand()*3)
    i = np.array([i]).T
    if rand == 0:
        class1 = np.hstack((class1, i))
    if rand == 1:
        class2 = np.hstack((class2, i))
    if rand == 2:
        class3 = np.hstack((class3, i))
        
for i in data3:
    rand = int(np.random.rand()*3)
    i = np.array([i]).T
    if rand == 0:
        class1 = np.hstack((class1, i))
    if rand == 1:
        class2 = np.hstack((class2, i))
    if rand == 2:
        class3 = np.hstack((class3, i))
```
