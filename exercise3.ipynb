Gradient Descent works by finding the direction of decreasing slope, and stepping in that direction iteratively. In order to implement this, I created a function func(w1, w2) which represents the function in the homework.

```python
import numpy as np

# Function from homework
def func(w1, w2):
    return (w1**2 + w2 -11)**2 + (w1 + w2**2 -7)**2


def gradientDesc(w, g):
    Wlist = []
    for i in range(0, 10000):
        # Compute current f(w)
        testf = func(w[0], w[1])
        Wlist.append(w.copy())
        # Compute step direction and change w accordingly
        if testf > func(w[0] + g, w[1]):
            w[0] = w[0] + g
        elif testf > func(w[0] - g, w[1]):
            w[0] = w[0] - g
        if testf > func(w[0], w[1] + g):
            w[1] = w[1] + g
        elif testf > func(w[0], w[1] - g):
            w[1] = w[1] - g
        # Test for convergence
        if testf - func(w[0], w[1]) < 10**-12:
            print("Final value of w: ", w)
            print("Number of iterations: ", i)
            return Wlist
```

Part 1:

```python
# Initalized w
w0 = [0, -4]
# stepsize
g = .005
W = gradientDesc(w0,g)
```

```python
# Initalized w
w0 = [0, -4]
# stepsize
g = .01
W = gradientDesc(w0,g)
```

The convergence speed of the gradient descent was faster with a larger step size since this meant that the function moved closer to the origin with each iteration. A larger step means faster movement toward the origin.

The solutions are not the same, these four different initial values converge to three different points which means each of these are local minimum

```python
from scipy.stats import multivariate_normal
import copy

u = np.array([-1, 0, 4, 1, 0.5])
I = np.eye(5)
covariance = 2*I

meanN = np.zeros([6,5])
eN = np.zeros(6)

for j in range(1, 7):
    xsum = np.zeros(5)
    N = 10**j
    for i in range(0, N - 1):
        x = np.array([multivariate_normal.rvs(u, covariance)])
        xsum = np.add(xsum, x)
    meanN[j-1] = xsum / N
    print("N = ", N)
    print("Mean:", meanN[j-1])
    eN[j-1] = np.linalg.norm(meanN[j-1] - u)**2
    print("Instantaneous error:", eN[j-1])
```

```python
powerN = np.linspace(1,6,6)
powerN = np.power(10, powerN)

plt.figure()
plt.loglog(powerN, eN)
plt.show()
```

The instantaneous error decreases exponentially as the number of samples increases exponentially.
