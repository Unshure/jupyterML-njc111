```python
import matplotlib.pyplot as plt
import numpy as np

x = [1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
y = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599]

plt.plot(x,y, 'ro--')
plt.title("Real US GPD (in trillions)")
plt.xlabel("Year")
plt.ylabel("Trillions of $")
```

Visually inspecting the graph above, it seems like the GPD is an exponential function of the Year. One way to find a rough estimate of this relationship would be to choose two points from the data and use those to come up with a relation.

I will use the following points in order to estimate the exponential function:

\begin{equation}
(y_{1},x_{1}) = (1930, 1.015)\quad (y_{2},x_{2}) =(2000, 13.131)
\end{equation}

\begin{equation}
y = A*B^x
\end{equation}

This can be represented as:

\begin{equation}
y_{1} = A*B^{x_{1}} \\
y_{2} = A*B^{x_{2}}
\end{equation}

To solve for A and B, we can do the following:

\begin{equation}
\dfrac{y_{1} = A*B^{x_{1}}}{y_{2} = A*B^{x_{2}}}
\end{equation}

\begin{equation}
\dfrac{y_{1}}{y_{2}} = B^{x_{1} - x_{2}}
\end{equation}

\begin{equation}
B = \dfrac{y_{1}}{y_{2}}^{{(x_{1} - x_{2})}^{-1}}
\end{equation}

From this we find that \begin{equation} B = 1.03725\end{equation}

Now we can find A by inserting B into the following equation

\begin{equation} A = \dfrac{y_{1}}{B^{x_{1}}} = 2.24738*10^{-31}\end{equation}

```python
import matplotlib.pyplot as plt
import numpy as np

x = [1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
y = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599]

t = np.arange(1920., 2010., 5)

plt.plot(x,y, 'ro--', t, (2.24738*10**-31) * (1.03725**t), 'b')
plt.title("Overlap of Estimated Equation")
plt.xlabel("Year")
plt.ylabel("Trillions of $")
```
